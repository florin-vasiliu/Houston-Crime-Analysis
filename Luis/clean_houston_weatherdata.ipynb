```python
import pandas as pd
import datetime as dt
```

```python
houston_weather = pd.read_csv('data/houston_weather_data.csv')
```

```python
houston_weather.head()
```

```python
houston_weather.count()
```

```python
houston_weather = houston_weather[["dt", "temp", "rain_1h", "weather_main", "weather_description"]]

houston_weather["rain_1h"] = houston_weather["rain_1h"].fillna(0)
```

```python
houston_weather
```

```python
houston_weather["Date_Time"] = pd.to_datetime(houston_weather["dt"], unit = 's')
houston_weather["Date"] = houston_weather["Date_Time"].dt.date
del houston_weather["dt"]
houston_weather.head()
```

```python
#create calculated fields which break out year, month, weeknum, day, etc
houston_weather["Year"] = pd.DatetimeIndex(houston_weather['Date_Time']).year
houston_weather["Month"] = pd.DatetimeIndex(houston_weather['Date_Time']).month
houston_weather["Week_Num"] = pd.DatetimeIndex(houston_weather['Date_Time']).week
houston_weather["Day"] = pd.DatetimeIndex(houston_weather['Date_Time']).day
houston_weather["Hour_Beginning"] = pd.DatetimeIndex(houston_weather['Date_Time']).hour
houston_weather["Day_of_Week"] = houston_weather['Date_Time'].dt.day_name()

houston_weather = houston_weather[['Date_Time', 'Date', 'Year', 'Month', 'Week_Num', 'Day', 'Hour_Beginning', 'Day_of_Week', 'temp', 'rain_1h', 'weather_main', 'weather_description']]
houston_weather
```

```python
#filter houston weather DF to 2018 - current
recent_houston_weather = houston_weather.loc[houston_weather["Year"] >= 2018]

recent_houston_weather.reset_index(drop=True, inplace=True)
```

```python
recent_houston_weather = recent_houston_weather.rename(columns={"temp":"Temperature", "rain_1h":"Precipitation", "weather_main":"Weather", "weather_description":"Weather_Description"})

recent_houston_weather
```

```python
#drop duplicates
recent_houston_weather.drop_duplicates(subset=['Year', 'Month', 'Day', 'Hour_Beginning'], keep='first', inplace=True)
recent_houston_weather
```

```python
recent_houston_weather.to_csv('houston_weather.csv')
```
