```python
import pandas as pd
```

```python
lat_lngs = ["LatLng1.csv", 
            "LatLng2.csv",
            "LatLng3.csv",
            "LatLng4.csv",
            "LatLng5.csv",
            "LatLng6.csv",
            "LatLng7.csv",
            "LatLng8.csv",]

li = []

for csv in lat_lngs:
    df = pd.read_csv(csv, index_col=None, header=0)
    li.append(df)

full_lat_lngs = pd.concat(li, axis=0, ignore_index=True)

full_lat_lngs
```

```python
crime_df = pd.read_csv('revised_crimes_geo.csv')

crime_df
```

```python
crime_df['Latitude'] = full_lat_lngs['Latitude']

crime_df
```

```python
crime_df['Longitude'] = full_lat_lngs['Longitude']

crime_df
```

```python
census_csvs = ["census1.csv", 
            "census2.csv",
           "census3.csv",
           "census4.csv",
           "census5.csv",
           "census6.csv",
           "census7.csv",
           "census8a.csv",
           "census8b.csv",
           "census8c.csv",
           "census8d.csv",
           "census8e.csv",
           "census8f.csv",]

li = []

for csv in census_csvs:
    df = pd.read_csv(csv, index_col=None, header=0)
    li.append(df)

tract_df = pd.concat(li, axis=0, ignore_index=True)

tract_df
```

```python
crime_df['Tract'] = tract_df['Census_Tract']

crime_df.head()
```

```python
cdf = crime_df[['Premise', 'Offense_Type', 'Offense_Count', 'Tract','Hour']]

cdf.dropna(axis=0, how='any', inplace=True)
cdf.head()
```

```python
for column in ['Premise', 'Offense_Type', 'Tract']:
    x = pd.api.types.CategoricalDtype(categories=list(cdf[column].unique()), ordered=False)
    cdf[column] = cdf[column].astype(x).cat.codes
# off_cat = pd.api.types.CategoricalDtype(categories=list(cdf.Offense_Type.unique()), ordered=False)
# cdf.Offense_Type.astype(off_cat)
```

```python
cdf.head()
```

```python
from sklearn.cluster import KMeans

import numpy as np

X = pd.DataFrame(data=[cdf.Premise.cat.codes, cdf.Offense_Type.cat.codes, cdf.Offense_Count, cdf.Tract.cat.codes,cdf.Hour]).to_numpy()
kmeans = KMeans(n_clusters=5, random_state=42, n_jobs=-1).fit(X)
kmeans.labels_
```

```python
kmeans.cluster_centers_
```
