```python
import pandas as pd
```

```python
# Clean 2020 Jan - May crime data

crime_2020_df = pd.read_csv("raw_data/2020_1-5_Jan-May.csv")
crime_2020_df = crime_2020_df.rename(columns={"Occurrence\nDate": "Date", 
                                              "Occurrence\nHour": "Hour",
                                              "NIBRS\nClass": "NIBRS_Class",
                                              "NIBRSDescription": "Offense_Type",
                                              "Offense\nCount": "Offense_Count",
                                              "Block Range": "Block_Range",
                                              "StreetName": "Street_Name",
                                              "Street\nType": "Street_Type",
                                              "ZIP Code": "Zip_Code"})
crime_2020_df.columns
```

```python
# Only take 2020 data we need
crime_2020_df = crime_2020_df[['Date','Hour','Offense_Type','Offense_Count','Premise','Block_Range','Street_Name','Street_Type','City']]
crime_2020_df
```

```python
# Clean 2019 Jan - Dec crime data

crime_2019_df = pd.read_csv("raw_data/2019_1-12_Jan-Dec.csv")
crime_2019_df = crime_2019_df.rename(columns={"Occurrence\nDate": "Date", 
                                              "Occurrence\nHour": "Hour",
                                              "NIBRS\nClass": "NIBRS_Class",
                                              "NIBRSDescription": "Offense_Type",
                                              "Offense\nCount": "Offense_Count",
                                              "Block Range": "Block_Range",
                                              "StreetName": "Street_Name",
                                              "Street\nType": "Street_Type",
                                              "ZIP Code": "Zip_Code"})
crime_2019_df.columns
```

```python
# Only take 2019 data we need
crime_2019_df = crime_2019_df[['Date','Hour','Offense_Type','Offense_Count','Premise','Block_Range','Street_Name','Street_Type','City']]
crime_2019_df['Offense_Type'].value_counts()
```

```python
# Clean 2018 Jan - May crime data
# Need to seperate 2018 data from first half and second half of the year due to different formatting of columns 

all_files = ["raw_data/2018_1_Jan.csv", 
            "raw_data/2018_2_Feb.csv",
            "raw_data/2018_3_Mar.csv",
            "raw_data/2018_4_Apr.csv",
            "raw_data/2018_5_May.csv"]

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

crime_2018_Jan_May_df = pd.concat(li, axis=0, ignore_index=True)
crime_2018_Jan_May_df = crime_2018_Jan_May_df.rename(columns={"Offense Type": "Offense_Type",
                                                              "Block Range": "Block_Range",
                                                              "Street Name": "Street_Name",
                                                              "Street\nType": "Street_Type",
                                                              "Type": "Street_Type",
                                                              "Offenses": "Offense_Count"})
crime_2018_Jan_May_df.columns
```

```python
# Only take 2018 Jan-May data we need
crime_2018_Jan_May_df = crime_2018_Jan_May_df[['Date','Hour','Offense_Type','Offense_Count','Premise','Block_Range','Street_Name','Street_Type']]
crime_2018_Jan_May_df['City'] = 'HOUSTON'
crime_2018_Jan_May_df
```

```python
# Clean 2018 June - Dec crime data

all_files = ["raw_data/2018_6_Jun.csv", 
            "raw_data/2018_7_Jul.csv",
            "raw_data/2018_8_Aug.csv",
            "raw_data/2018_9_Sep.csv",
            "raw_data/2018_10_Oct.csv",
            "raw_data/2018_11_Nov.csv",
            "raw_data/2018_12_Dec.csv"]

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

crime_2018_Jun_Dec_df = pd.concat(li, axis=0, ignore_index=True)
crime_2018_Jun_Dec_df = crime_2018_Jun_Dec_df.rename(columns={"Occurrence Date": "Date",
                                                              "Occurrence Hour": "Hour",
                                                              "NIBRS Description": "Offense_Type",
                                                              "Offense Count": "Offense_Count",
                                                              "Block Range": "Block_Range",
                                                              "Street Name": "Street_Name",
                                                              "Street Type": "Street_Type"})
crime_2018_Jun_Dec_df.columns
```

```python
# Only take 2018 Jun-Dec data we need
crime_2018_Jun_Dec_df = crime_2018_Jun_Dec_df[['Date','Hour','Offense_Type','Offense_Count','Premise','Block_Range','Street_Name','Street_Type']]
crime_2018_Jun_Dec_df['City'] = 'HOUSTON'
crime_2018_Jun_Dec_df
```

```python
clean_crime_data_df = pd.concat([crime_2020_df, 
                                 crime_2019_df,
                                 crime_2018_Jan_May_df, 
                                 crime_2018_Jun_Dec_df,], ignore_index=True)
```

```python
relevant_crimes = ['Theft of motor vehicle parts or accessory', 'Aggravated Assault',
       'Destruction, damage, vandalism', 'Simple assault',
       'Theft from motor vehicle',
       'Intimidation', 'Robbery', 'Forcible rape', 'Motor vehicle theft',
       'All other larceny',
       'Statutory rape',
       'Burglary, Breaking and Entering', 'Murder, non-negligent',
       'Arson', 'Pocket-picking',
       'Theft from building', 'Forcible fondling', 'Purse-snatching',
       'False pretenses, swindle', 'Stolen property offenses',
       'Human Trafficking/Commercial Sex Act',
       'Kidnapping, abduction',
       'Sexual assault with an object','Negligent manslaughter',
       'Human Trafficking/Involuntary Servitude', 'Theft', 'Burglary',
       'AutoTheft', 'Rape', 'Murder']

crime_df = clean_crime_data_df[clean_crime_data_df['Offense_Type'].isin(relevant_crimes)]
crime_df
```

```python
theft = ['Theft of motor vehicle parts or accessory', 'Theft from motor vehicle', 'Purse-snatching', 'Stolen property offenses',
        'Motor vehicle theft', 'Theft from building', 'Pocket-picking', 'AutoTheft', 'All other larceny', 'False pretenses, swindle',]

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(theft)] = 'Theft'

assault = ['Aggravated Assault', 'Simple assault']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(assault)] = 'Assault'

burglary = ['Burglary, Breaking and Entering', 'Robbery', 'Burglary']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(burglary)] = 'Burglary/Robbery'

sex_crimes = ['Forcible rape','Statutory rape','Forcible fondling','Human Trafficking/Commercial Sex Act',
             'Sexual assault with an object','Rape']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(sex_crimes)] = 'Sex Crime'

vandalism = ['Destruction, damage, vandalism']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(vandalism)] = 'Vandalism'

murder = ['Murder, non-negligent','Negligent manslaughter','Murder']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(murder)] = 'Murder/Manslaughter'

intimidation = ['Intimidation']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(intimidation)] = 'Intimidation'

kidnapping = ['Kidnapping, abduction','Human Trafficking/Involuntary Servitude']

crime_df['Offense_Type'][crime_df['Offense_Type'].isin(kidnapping)] = 'Kidnapping'
```

```python
address_list = clean_crime_data_df['Street_Address'].to_list()
city_list = clean_crime_data_df['City'].to_list()

print(len(address_list))
print(len(city_list))
```

```python
import requests
import csv

census_tract = []

for street, city in zip(address_list[:7500], city_list[:7500]):
    try:
        census_url = f'https://geocoding.geo.census.gov/geocoder/geographies/address?street={street}&city={city}&state=TX&benchmark=Public_AR_Census2010&vintage=Census2010_Census2010&layers=14&format=json'
        response = requests.get(census_url)
        census_json = response.json()
        census_tract.append(census_json['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT'])
        print(f"{street} Census Tract: {census_json['result']['addressMatches'][0]['geographies']['Census Blocks'][0]['TRACT']}")
        print('-------------')                                        
    except:
        print(f'{street}: has failed you')
        census_tract.append('NaN')

test_df = pd.DataFrame(census_tract, columns=['Census_Tract'])

test_df.to_csv('census.csv', index=False)
```
