# Dataset 1: API call from census.gov

```python
#Dataset:
#https://api.census.gov/data.html
#American Community Survey (ACS) 5-Year Data Profiles year 2018 

base_url="https://api.census.gov/data"
year="2018"

#api.census.gov/data.html
dataset_name="acs/acs5/subject"

#Selected the variables:
#GEO_ID
#S0101_C01_001E - Estimate!!Total!!Total population
#S1101_C01_001E - Estimate!!Total!!HOUSEHOLDS!!Total households
#S1901_C01_012E - Estimate!!Households!!Median income (dollars)

variable="GEO_ID,S0101_C01_001E,S1101_C01_001E,S1901_C01_012E"
geography="for=tract:*&in=state:48"
```

```python
#define URL for querying census data
url=f"{base_url}/{year}/{dataset_name}?get={variable}&{geography}&key={census_api_key}"
#get response from census and store it
response = requests.get(url)
response.json()
```

```python
#store response into dataframe
census_result_df=pd.DataFrame(response.json(), columns=response.json()[0])

#remove duplicate header row
census_result_df=census_result_df.drop(axis=0,index=0)

#rename columns 
census_result_df=census_result_df.rename(columns={"S0101_C01_001E":"Total_Popupation",
                                                  "S1101_C01_001E":"Total_Households",
                                                  "S1901_C01_012E":"Median_Household_Income",
                                                  "state":"state_fips",
                                                  "county":"county_fips",
                                                  "tract":"tract_code"
                                                 })

#set index to tract_code and sort 
census_result_df=census_result_df.set_index("tract_code")#.sort_values(by="tract_code",ascending=True)

census_result_df
```

```python
#write results to csv file
census_result_df.to_csv("census_result.csv")
```

```python
#Example of request at https://geocoding.geo.census.gov/
number = "100"
street = "bryan"
address_url = "https://geocoding.geo.census.gov/geocoder/geographies/address" \
    +f"?street={number}+{street}" \
    +"&city=Houston&state=TX&zip=&benchmark=4&vintage=4&format=json"
address_eg = requests.get (address_url)
```

```python
#initiate lists to store requested results
matched_address = []
x_coord = []
y_coord = []
census_tract = []

#iterate only the first 5 rows of the crime_df

for index, row in crime_df.head().iterrows():
    try:
        #build query and perform request
        number = row["Block Range"].split("-")[0]
        street = row["Street Name"]
        address_url = "https://geocoding.geo.census.gov/geocoder/geographies/address" \
            +f"?street={number}+{street}" \
            +"&city=Houston&state=TX&zip=&benchmark=4&vintage=4&format=json"
        result_json = requests.get(address_url).json()

        #write results into lists
        matched_address.append(result_json["result"]['addressMatches'][0]['matchedAddress'])
        x_coord.append(result_json["result"]['addressMatches'][0]['coordinates']["x"])
        y_coord.append(result_json["result"]['addressMatches'][0]['coordinates']["y"])
        census_tract.append(result_json["result"]['addressMatches'][0]['geographies']['Census Tracts'][0]['TRACT'])
    except:
        #write not found values
        matched_address.append("NotFound")
        x_coord.append("NaN")
        y_coord.append("NaN")
        census_tract.append("NaN")     
```
